```python

def convert_people_cell(cell):
    if cell == "n.a.":
        return "sam walton"
    return cell

def convert_aps_cell(cell):
    if cell == "not available":
        return None 
    return cell

pdf = pd.read_excel("stock_data.xlsx","Sheet1",converters = {
    "people": convert_people_cell,
    "eps": convert_aps_cell
    })
pdf 
```

```python
pdf.to_excel("cleaned.xlsx",sheet_name = "stocks",index = False)
```
